```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io
sns.set(style="whitegrid")  # set style

data = io.StringIO(""""Country"  2016  2019
"McDonalds" 36.389 40.413 
"Starbucks" 14.795 21.550
"CHICK_FIL_A" 7.973 11.000
"Taco_Bell" 9.353 11.000
"Burger_King" 9.749 11.300
"Subway" 11.300 9.000
"WENDYS" 9.510 10.565
"DUNKIN_DONUTS" 8.200 9.220
"DOMINOS" 5.335 7.100
"Panera_Bread" 5.200 6.525""")

df = pd.read_csv(data, sep="\s+", quotechar='"')
df = df.set_index("Country").sort_values("2019")
df["change"] = df["2019"] / df["2016"] - 1


plt.figure(figsize=(14,10))
y_range = np.arange(1, len(df.index) + 1)
colors = np.where(df['2019'] > df['2016'], '#d9d9d9', '#d57883')
plt.hlines(y=y_range, xmin=df['2016'], xmax=df['2019'],
           color=colors, lw=20)
plt.scatter(df['2016'], y_range, color='#0096d7', s=250, label='2016', zorder=3)
plt.scatter(df['2019'], y_range, color='#003953', s=250 , label='2019', zorder=3)
for (_, row), y in zip(df.iterrows(), y_range):
    plt.annotate(f"{row['change']:+.0%}" ,(max(row["2016"], row["2019"]) + 4, y - 0.25), fontsize= 25)
plt.legend(ncol=7, bbox_to_anchor=(1., 1.01), loc="lower right", frameon=True, fontsize= 25)

plt.xlabel('Sales in Millions', fontsize= 25) 
plt.ylabel('Fast Food Company', fontsize= 25) 
plt.yticks(y_range, df.index, fontsize= 25)
plt.title("Sales growth: Top fastfood restaurants from 2016-2019",fontsize=20 ,loc='left')
plt.xticks((range(3,50,5)),fontsize= 25)
plt.gcf().subplots_adjust(left=0.35)
plt.tight_layout()
plt.show()
```
